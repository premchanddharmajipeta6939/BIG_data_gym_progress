# Day-to-day Weight Change

```python
%sql
SELECT
  Day_Number,
  Weight_kg,
  LAG(Weight_kg) OVER (ORDER BY Day_Number)       AS Prev_Weight,
  Weight_kg - LAG(Weight_kg) OVER (ORDER BY Day_Number) AS Weight_Change
FROM workspace.gym_progress.gym_progress_clean
ORDER BY Day_Number;
```

#Cumulative Steps (Running Total)

```python
%sql
SELECT
  Day_Number,
  Steps_Walked,
  SUM(Steps_Walked) OVER (ORDER BY Day_Number) AS Cumulative_Steps
FROM workspace.gym_progress.gym_progress_clean
ORDER BY Day_Number;
```

#3-Day Moving Average of Weight

```python
%sql
SELECT
  Day_Number,
  Weight_kg,
  AVG(Weight_kg) OVER (
      ORDER BY Day_Number
      ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
  ) AS Moving_Avg_Weight_3d
FROM workspace.gym_progress.gym_progress_clean

ORDER BY Day_Number;
```

#Rank Days by Calorie Surplus (RANK)

```python
%sql
SELECT
  Day_Number,
  Calories_Intake,
  Active_Calories,
  Calorie_Balance,
  RANK() OVER (ORDER BY Calorie_Balance DESC) AS Surplus_Rank
FROM workspace.gym_progress.gym_progress_clean
ORDER BY Surplus_Rank;
```
